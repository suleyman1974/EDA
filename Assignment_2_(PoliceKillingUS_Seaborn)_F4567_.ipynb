## 2: Plot 12 most dangerous cities

```python
kill.city.value_counts().head(12)
```

```python
kill['city'].value_counts().values[:12]
```

```python
kill['city'].value_counts().index[:12]
```

```python
# with seaborn barplot
fig, ax = plt.subplots(figsize=(10,6))
sns.barplot(x = kill.city.value_counts().index[:12], y = kill.city.value_counts().values[:12])
ax.set_xticklabels(kill.city.value_counts().index[:12], rotation =45)
ax.set_title("Most Dangereous City", color = "k", fontsize=14)
plt.show()
```

```python
kill.city.value_counts()[:13]
```

```python
# with seaborn countplot

fig, ax = plt.subplots(figsize=(10,6))
sns.countplot(x = 'city', data = kill, order = kill.city.value_counts().index[:12])
ax.set_xticklabels(kill.city.value_counts().index[:12], rotation =45)
ax.set_title("Most Dangereous City", color = "k", fontsize=14)
plt.show()
```
